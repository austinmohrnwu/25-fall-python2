```python
import math
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y      
    def __repr__(self):
        return f"Point({self.x}, {self.y})"
    def distance(self, other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)
class Polygon:
    def __init__(self, * points):
        self.points = list(points)
    def centroid(self):
        n = len(self.points)
        avg_x = sum(p.x for p in self.points) / n
        avg_y = sum(p.y for p in self.points) / n
        return Point(avg_x, avg_y)
    def perimeter(self):
        total = 0
        n = len(self.points)
        for i in range(n):
            total += self.points[i].distance(self.points[(i+1) % n])
        return total
    def is_regular(self):
        n = len(self.points)
        sides = []
        for i in range(n):
            sides.append(self.points[i].distance(self.points[(i+1) % n]))
        first = sides[0]
        return all(abs(s - first) < 1e-9 for s in sides)
class Triangle(Polygon):
    pass
class Quadrilateral(Polygon):
    pass
```

```python
tri = Triangle(Point(0, 0), Point(1, 3), Point(2, 0))
tri.centroid()
```

```python
quad = Quadrilateral(Point(0, 0), Point(1, 0), Point(1, 1), Point(0, 1))
quad.perimeter()
```

```python
quad = Quadrilateral(Point(0, 0), Point(1, 0), Point(1, 1), Point(0, 1))
quad.is_regular()
```
