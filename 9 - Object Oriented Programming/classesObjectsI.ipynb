```python
class Point:
    '''
    Represents a point in 2D space
    attributes: x (float), y (float)
    '''
    def __init__(self, input_x = 0, input_y = 0):
        self.x = input_x
        self.y = input_y
class Rectangle:
    '''
    Represents a rectangle with lower left corner at corner and specified height and width.
    attributes: corner (Point object), height (float), width (float)
    '''
    def __init__(self, input_corner = Point(), input_height = 0, input_width = 0):
        self.corner = input_corner
        self.height = input_height
        self.width = input_width
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
class Rectangle:
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height
def get_corners(rect):
    top_left = Point(rect.x, rect.y)
    top_right = Point(rect.x + rect.width, rect.y)
    bottom_left = Point(rect.x, rect.y + rect.height)
    bottom_right = Point(rect.x + rect.width, rect.y + rect.height)
    return [top_left, top_right, bottom_left, bottom_right]
```

```python
r = Rectangle(10, 20, 50, 30)
corners = get_corners(r)
for i, c in enumerate(corners, start=1):
    print(f"Corner {i}: ({c.x}, {c.y})")
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
class Circle:
    def __init__(self, center, radius):
        """
        Initialize a Circle object.
        Parameters:
        center (Point): The center of the circle.
        radius (float): The radius of the circle.
        """
        self.center = center
        self.radius = radius
```

```python
import math
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
class Circle:
    def __init__(self, center, radius):
        self.center = center
        self.radius = radius
def distance(p1, p2):
    """Return the distance between two Point objects."""
    dx = p2.x - p1.x
    dy = p2.y - p1.y
    return math.sqrt(dx**2 + dy**2)
def point_in_circle(circle, point):
    """Return True if the point is inside or on the circle, False otherwise."""
    return distance(circle.center, point) <= circle.radius
```

```python
c = Circle(Point(0, 0), 5)
p1 = Point(3, 4) 
p2 = Point(6, 0)
print(point_in_circle(c, p1))
print(point_in_circle(c, p2)) 
```

```python

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
class Rectangle:
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height
class Circle:
    def __init__(self, center, radius):
        self.center = center
        self.radius = radius
import math
def distance(p1, p2):
    dx = p2.x - p1.x
    dy = p2.y - p1.y
    return math.sqrt(dx**2 + dy**2)
def get_corners(rect):
    top_left = Point(rect.x, rect.y)
    top_right = Point(rect.x + rect.width, rect.y)
    bottom_left = Point(rect.x, rect.y + rect.height)
    bottom_right = Point(rect.x + rect.width, rect.y + rect.height)
    return [top_left, top_right, bottom_left, bottom_right]
def point_in_circle(circle, point):
    return distance(circle.center, point) <= circle.radius
def rect_in_circle(circle, rect):
    """Return True if all four corners of the rectangle lie inside or on the circle."""
    corners = get_corners(rect)
    return all(point_in_circle(circle, corner) for corner in corners)
```

```python
rect1 = Rectangle(1, 2, 4, 3)
circle1 = Circle(Point(1, 2), 1)
print(rect_in_circle(circle1, rect1))
```

```python
rect2 = Rectangle(0, 0, 1, 1)
circle2 = Circle(Point(1, 1), 5)
print(rect_in_circle(circle2, rect2))
```
