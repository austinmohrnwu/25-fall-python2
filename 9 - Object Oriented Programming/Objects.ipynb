```python
# Write a function named get_corners that takes a Rectangle object as input. It should return a list of Point objects: 
# one for each corner of the Rectangle.

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Rectangle:
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height

def get_corners(rect):
    top_left = Point(rect.x, rect.y)
    top_right = Point(rect.x + rect.width, rect.y)
    bottom_left = Point(rect.x, rect.y + rect.height)
    bottom_right = Point(rect.x + rect.width, rect.y + rect.height)
    
    return [top_left, top_right, bottom_left, bottom_right]
```

```python
r = Rectangle(10, 20, 50, 30)
corners = get_corners(r)

for i, c in enumerate(corners, start=1):
    print(f"Corner {i}: ({c.x}, {c.y})")
```

```python
# Write a function named point_in_circle that takes a Circle object and a Point object as inputs. 
# The function returns True if the Point is inside of or on the boundary of the Circle and False otherwise. 
# (Devise a way to use your distance function to help define point_in_circle.)
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Circle:
    def __init__(self, center, radius):
        self.center = center
        self.radius = radius

def distance(p1, p2):
    """Return the distance between two Point objects."""
    dx = p2.x - p1.x
    dy = p2.y - p1.y
    return math.sqrt(dx**2 + dy**2)

def point_in_circle(circle, point):
    """Return True if the point is inside or on the circle, False otherwise."""
    return distance(circle.center, point) <= circle.radius
```

```python
c = Circle(Point(0, 0), 5)
p1 = Point(3, 4)  # On the circle
p2 = Point(6, 0)  # Outside the circle

print(point_in_circle(c, p1))
print(point_in_circle(c, p2)) 
```

```python
# Write a function named rect_in_circle that takes a Circle object and a Rectangle object as inputs. 
# The function returns True if all four corners of the Rectangle lie inside of or on the boundary of the circle. 
# (Devise a way to use your get_corners and point_in_circle functions to help define rect_in_circle.)
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Rectangle:
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height

class Circle:
    def __init__(self, center, radius):
        self.center = center
        self.radius = radius

import math

def distance(p1, p2):
    dx = p2.x - p1.x
    dy = p2.y - p1.y
    return math.sqrt(dx**2 + dy**2)

def get_corners(rect):
    top_left = Point(rect.x, rect.y)
    top_right = Point(rect.x + rect.width, rect.y)
    bottom_left = Point(rect.x, rect.y + rect.height)
    bottom_right = Point(rect.x + rect.width, rect.y + rect.height)
    return [top_left, top_right, bottom_left, bottom_right]

def point_in_circle(circle, point):
    return distance(circle.center, point) <= circle.radius

def rect_in_circle(circle, rect):
    """Return True if all four corners of the rectangle lie inside or on the circle."""
    corners = get_corners(rect)
    return all(point_in_circle(circle, corner) for corner in corners)
```

```python
c = Circle(Point(0, 0), 10)
r1 = Rectangle(-3, -3, 6, 6)   # Fully inside the circle
r2 = Rectangle(-8, -8, 10, 10) # Partially outside the circle

print(rect_in_circle(c, r1))  # True
print(rect_in_circle(c, r2))  # False
```
