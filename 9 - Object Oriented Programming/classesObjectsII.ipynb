```python
import math
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    def __str__(self):
        return f"({self.x}, {self.y})"
class Circle:
    def __init__(self, center=Point(), radius=1):
        self.center = center
        self.radius = radius
    def __str__(self):
        return f"Circle(center={self.center}, radius={self.radius:.2f})"
    def get_area(self):
        return math.pi * (self.radius ** 2)
    def to_square(self):
        """Return a square Rectangle with the same area and center."""
        side = math.sqrt(self.get_area())
        lower_left_x = self.center.x - side / 2
        lower_left_y = self.center.y - side / 2
        return Rectangle(Point(lower_left_x, lower_left_y), side, side)
class Rectangle:
    def __init__(self, corner=Point(), width=1, height=1):
        self.corner = corner
        self.width = width
        self.height = height
    def __str__(self):
        return f"Rectangle(corner={self.corner}, width={self.width:.2f}, height={self.height:.2f})"
    def get_area(self):
        return self.width * self.height
    def __lt__(self, other):
        """Overload < operator based on area."""
        return self.get_area() < other.get_area()
    def to_circle(self):
        """Return a Circle with the same center and area."""
        center_x = self.corner.x + self.width / 2
        center_y = self.corner.y + self.height / 2
        center = Point(center_x, center_y)
        area = self.get_area()
        radius = math.sqrt(area / math.pi)
        return Circle(center, radius)
```

```python
rect = Rectangle(Point(), 2, 2)
circle_from_rect = rect.to_circle()
print(circle_from_rect)
```

```python
circ = Circle(Point(), 1)
rect_from_circ = circ.to_square()
print(rect_from_circ)
```

```python
r1 = Rectangle(Point(), 1, 1)
r2 = Rectangle(Point(), 1, 2)
print(r1 < r2)
```
